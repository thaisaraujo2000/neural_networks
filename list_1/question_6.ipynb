```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    fx = sigmoid(x)
    return fx * (1 - fx)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    fx = np.tanh(x)
    return 1 - fx**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x <= 0, 0, 1)
```

```python
# Definindo o intervalo de valores de v (potencial de ativa√ß√£o)
x = np.linspace(-5, 5, 100)

# Criando figura e eixos
fig, axs = plt.subplots(2, 3, figsize=(12, 6), constrained_layout=True)

# Plotando os gr√°ficos das fun√ß√µes de ativa√ß√£o e suas derivadas
axs[0, 0].plot(x, sigmoid(x), color='#9B4F96', linewidth=2)
axs[0, 0].set_title('Sigmoid')
axs[0, 0].set_xlim([-5, 5])
axs[0, 0].set_ylim([-0.2, 1.2])
axs[0, 0].set_xticks([-5, 0, 5])
axs[0, 0].set_xticklabels(['$-5$', '$0$', '$5$'])
axs[0, 0].set_yticks([0, 0.5, 1])
axs[0, 0].set_yticklabels(['$0$', '$0.5$', '$1$'])
axs[0, 0].set_facecolor('#F7F7F7')
axs[0, 0].grid(True, alpha=0.5)

axs[1, 0].plot(x, sigmoid_derivative(x), color='#E45756', linewidth=2)
axs[1, 0].set_title('Sigmoid Derivative')
axs[1, 0].set_xlim([-5, 5])
axs[1, 0].set_ylim([-0.5, 0.5])
axs[1, 0].set_xticks([-5, 0, 5])
axs[1, 0].set_xticklabels(['$-5$', '$0$', '$5$'])
axs[1, 0].set_yticks([-0.5, 0, 0.5])
axs[1, 0].set_yticklabels(['$-0.5$', '$0$', '$0.5$'])
axs[1, 0].set_facecolor('#F7F7F7')
axs[1, 0].grid(True, alpha=0.5)

axs[0, 1].plot(x, tanh(x), color='#1CA1C1', linewidth=2)
axs[0, 1].set_title('Tanh')
axs[0, 1].set_xlim([-5, 5])
axs[0, 1].set_ylim([-1.2, 1.2])
axs[0, 1].set_xticks([-5, 0, 5])
axs[0, 1].set_xticklabels(['$-5$', '$0$', '$5$'])
axs[0, 1].set_yticks([-1, 0, 1])
axs[0, 1].set_yticklabels(['$-1$', '$0$', '$1$'])
axs[0, 1].set_facecolor('#F7F7F7')
axs[0, 1].grid(True, alpha=0.5)

axs[1, 1].plot(x, tanh_derivative(x), color='#4CB944', linewidth=2)
axs[1, 1].set_title('Tanh Derivative')

axs[1, 1].grid(color='gray', linestyle='--', linewidth=0.5, axis='both', alpha=0.5)

axs[0, 2].plot(x, relu(x), color='tab:red', linewidth=2)
axs[0, 2].set_title('ReLU')
axs[0, 2].grid(color='gray', linestyle='--', linewidth=0.5, axis='both', alpha=0.5)

axs[1, 2].plot(x, relu_derivative(x), color='tab:red', linewidth=2)
axs[1, 2].set_title('ReLU Derivative')
axs[1, 2].grid(color='gray', linestyle='--', linewidth=0.5, axis='both', alpha=0.5)

plt.tight_layout()
plt.show()
```

Podemos observar que a fun√ß√£o sigmoide e a tangente hiperb√≥lica s√£o suaves e diferenci√°veis em toda a sua extens√£o, enquanto a fun√ß√£o ReLu √© n√£o-linear e n√£o suave em $v=0$, mas √© diferenci√°vel em todos os outros lugares.

As derivadas da fun√ß√£o sigmoide e da tangente hiperb√≥lica s√£o m√°ximas em $v=0$, enquanto a derivada da fun√ß√£o ReLu √© zero em $v\leq0$ e 1 em $v\geq0$.

Essas diferen√ßas nas fun√ß√µes de ativa√ß√£o e suas derivadas podem afetar o desempenho da rede neural em diferentes tarefas e arquiteturas. A escolha da fun√ß√£o de ativa√ß√£o adequada depende do problema em quest√£o e da arquitetura da rede neural utilizada.

Por exemplo, em tarefas de classifica√ß√£o bin√°ria, a fun√ß√£o sigmoide √© comumente usada na camada de sa√≠da, enquanto a ReLu √© usada nas camadas ocultas em redes profundas devido ao seu desempenho em evitar o problema de gradiente desaparecendo em redes profundas.

J√° a fun√ß√£o tangente hiperb√≥lica pode ser usada em redes neurais para problemas de regress√£o ou classifica√ß√£o, principalmente quando os dados possuem uma distribui√ß√£o normal.

Para a fun√ß√£o sigmoide, temos que:
$\varphi(v) = \frac{1}{1 + e^{-av}}$

Onde "a" √© um fator de escala que determina a "curvatura" da fun√ß√£o sigmoide.
Para calcular a derivada de $\varphi(v)$, podemos usar a regra da cadeia:

\begin{align*}
\varphi'(v) &= \frac{d}{dv} \left[\frac{1}{1 + e^{-av}}\right] \
&= \varphi(v)^2 \cdot \frac{d}{dv} \left[\frac{1}{\varphi(v)}\right] \cdot \frac{d}{dv} \left[1 + e^{-av}\right] \
&= \varphi(v)^2 \cdot \left[-\frac{1}{\varphi(v)^2}\right] \cdot e^{-av} \cdot (-a) \
&= a\varphi(v) \cdot (1-\varphi(v))
\end{align*}

Portanto, $\varphi'(v) = a\varphi(v) \cdot (1-\varphi(v))$ para a fun√ß√£o sigmoide.

J√° para a fun√ß√£o tangente hiperb√≥lica, temos que:

$\varphi(v) = \tanh(av) = \frac{e^{av} - e^{-av}}{e^{av} + e^{-av}}$

Onde "a" √© um fator de escala que determina a "curvatura" da fun√ß√£o tangente hiperb√≥lica.

Para calcular a derivada de $\varphi(v)$, podemos usar a regra da cadeia:

$\varphi'(v) = \frac{d}{dv}[\tanh(av)]$

$= \mathrm{sech}^2(av) \cdot a$

$= \frac{1}{\cosh^2(av)} \cdot a$

$= \frac{1}{(1 + \tanh^2(av))} \cdot a$

$= \frac{1}{(1 + \varphi(v)^2)} \cdot a$

$= \frac{a}{2} \cdot \frac{2}{(1 + \varphi(v)^2)} $

$= \frac{a}{2} \cdot (1 - \frac{2}{(1 + \varphi(v)^2)})$

$= \frac{a}{2} \cdot (1 - \varphi(v)^2)$

Portanto, $\varphi'(v) = \frac{a}{2} \cdot (1 - \varphi(v)^2)$ para a fun√ß√£o tangente hiperb√≥lica.
